```python
file_path = '../data/raw/consolidation_etalab_schema_irve_statique_v_231_20250804_prepared.csv'
df = pd.read_csv(file_path)
print(df.columns)

# Vérifier les doublons dans l'ensemble du DataFrame
duplicates = df[df.duplicated()]

# Afficher les doublons
print("Nombre de doublons trouvés :", len(duplicates))
if not duplicates.empty:
    print("Exemples de doublons :")
    print(duplicates.head())
else:
    print("Aucun doublon trouvé.")

# Optionnel : Vérifier les doublons basés sur certaines colonnes spécifiques
# Par exemple, vérifier les doublons basés sur 'nom_station' et 'adresse_station'
specific_columns_duplicates = df[df.duplicated(subset=['nom_station', 'adresse_station'])]

print("\nNombre de doublons basés sur des colonnes spécifiques :", len(specific_columns_duplicates))
if not specific_columns_duplicates.empty:
    print("Exemples de doublons basés sur des colonnes spécifiques :")
    print(specific_columns_duplicates.head())
else:
    print("Aucun doublon trouvé basé sur des colonnes spécifiques.")
```

```python
# Calculer le pourcentage de valeurs manquantes par colonne
missing_percentage = df.isnull().mean() * 100

# Filtrer les colonnes avec des valeurs manquantes
missing_columns = missing_percentage[missing_percentage > 0]

# Afficher le pourcentage de valeurs manquantes uniquement pour les colonnes concernées
print("Pourcentage de valeurs manquantes par colonne :")
print(missing_columns)
```

```python
# Remplacer les valeurs manquantes dans chaque colonne
df['nom_amenageur'].fillna('Inconnu', inplace=True)
df['contact_amenageur'].fillna('Non spécifié', inplace=True)
df['nom_operateur'].fillna('Inconnu', inplace=True)
df['id_station_local'].fillna(-1, inplace=True)
df['adresse_station'].fillna('Adresse inconnue', inplace=True)
df['id_pdc_local'].fillna(-1, inplace=True)
df['gratuit'].fillna('Non spécifié', inplace=True)
df['paiement_cb'].fillna('Non spécifié', inplace=True)
df['paiement_autre'].fillna('Non spécifié', inplace=True)
df['tarification'].fillna('Non spécifié', inplace=True)
df['restriction_gabarit'].fillna('Aucune restriction', inplace=True)
df['raccordement'].fillna('Non spécifié', inplace=True)
df['date_mise_en_service'].fillna(pd.Timestamp('1970-01-01'), inplace=True)
df['date_maj'].fillna(pd.Timestamp('1970-01-01'), inplace=True)
df['cable_t2_attache'].fillna('Non spécifié', inplace=True)

# Sauvegarder le DataFrame nettoyé dans un nouveau fichier CSV
cleaned_file_path = '../data/clean/borneClean.csv'
df.to_csv(cleaned_file_path, index=False)
```

chech fichier clean

```python
# Charger le fichier CSV nettoyé
cleaned_file_path = '../data/clean/borneClean.csv'
df_cleaned = pd.read_csv(cleaned_file_path)

# Vérification des valeurs manquantes
missing_values = df_cleaned.isnull().sum()
print("Valeurs manquantes par colonne après nettoyage :")
print(missing_values)

# Vérification des doublons
duplicates = df_cleaned[df_cleaned.duplicated()]
print("\nNombre de doublons après nettoyage :", len(duplicates))

# Validation des types de données
print("\nTypes de données des colonnes :")
print(df_cleaned.dtypes)

# Analyse statistique descriptive
print("\nStatistiques descriptives :")
print(df_cleaned.describe(include='all'))

# Vérification des valeurs remplacées
print("\nValeurs uniques dans les colonnes remplacées :")
print("Colonne 'date_mise_en_service' :", df_cleaned['date_mise_en_service'].unique()[:5], "...")
print("Colonne 'date_maj' :", df_cleaned['date_maj'].unique()[:5], "...")
print("Colonne 'cable_t2_attache' :", df_cleaned['cable_t2_attache'].unique())
```
